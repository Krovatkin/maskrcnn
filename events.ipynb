```python
import json
```

```python
events = json.load(open("maskrcnn.json"))['traceEvents']
```

```python
events[0]
```

```python
import pprint

def matches(pats, haystack):
    for p in pats:
        if p in haystack:
            return True

    return False

in_dzone = False
devents = []
for e in events:
    if matches(['nonzero', 'unique'], e['name']):
        in_dzone = True

    if matches(['sum', 'min', 'max', 'mean'], e['name']):
        if len(devents) > 0:
            print("==========DYNAMIC ZONE=============")
            pprint.pprint(devents)
            devents.clear()
        in_dzone = False
    
    if in_dzone:
        
        devents.append(e['name'])

if len(devents) > 0:
    print("==========DYNAMIC ZONE=============")
    pprint.pprint(devents)


```
